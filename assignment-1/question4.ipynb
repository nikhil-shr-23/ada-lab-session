```python
#QUESTION-4

import time
import pandas as pd

# T(n) = T(n/2) + n
c1 = 0
def rec1(n):
    global c1
    c1 += 1
    if n <= 1:
        return 1
    return rec1(n // 2) + n

# T(n) = 2T(n/2) + n
c2 = 0
def rec2(n):
    global c2
    c2 += 1
    if n <= 1:
        return 1
    return rec2(n // 2) + rec2(n // 2) + n

# input sizes
sizes = [8, 16, 32, 64, 128]

rows = []

for n in sizes:
    # Recurrence 1
    c1 = 0
    s = time.time()
    rec1(n)
    t1 = time.time() - s

    # Recurrence 2
    c2 = 0
    s = time.time()
    rec2(n)
    t2 = time.time() - s

    rows.append([
        n,
        t1, c1,
        t2, c2
    ])

#resutl table
df = pd.DataFrame(
    rows,
    columns=[
        "n",
        "T(n)=T(n/2)+n Time", "Calls",
        "T(n)=2T(n/2)+n Time", "Calls"
    ]
)

df
```
